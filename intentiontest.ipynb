```python
import requests
import pandas as pd
import numpy as np
import time
```

```python
sheet = pd.read_excel("南网测试语料.xlsx")
```

```python
intention = []
faq_intention = []
for tmp in sheet.get("语料"):
    if(tmp is np.nan):
        intention.append('')
        continue
    else:
        result = requests.get("http://101.200.240.22:8078/slu/parse/test",params={'companyId':"uc6bf8acdc3e94ff9a1f10700a07205e7","inputType":"text","inputValue":tmp,"label":"0","intentions":"","lastReplyString":"","keypadParam":"","mustFlowName":""})
        print("语料："+ str (tmp) + "结果：")
        print(result.json())
        intention.append("")
        faq_intention.append("")
        for result_tmp in result.json():
            if result_tmp.get('type') == 'stateNode':
                intention[-1] = result_tmp.get('results')
            elif result_tmp.get('type') == 'faq':
                faq_intention[-1] = result_tmp.get('results')
            else:
                continue
        time.sleep(2)
```

```python
def gen_id2corpus(corpus_file):
    id2corpus = []
    with open(corpus_file, "r", encoding="utf-8") as f:
        for idx, line in enumerate(f):
            id2corpus.append( line.rstrip())
    return id2corpus
data_corpus = gen_id2corpus("clusterKnn/corpus.csv")
```
